```python
from glob import glob
import pyterrier as pt
from tira.third_party_integrations import ensure_pyterrier_is_loaded
```

```python
ensure_pyterrier_is_loaded()
ds_id = "radboud-validation-20251114-training"
dataset = pt.datasets.get_dataset(f"irds:ir-lab-wise-2025/{ds_id}")
topics = dataset.get_topics("title")
qrels = dataset.get_qrels()
```

```python
runs = {}

for run_name in glob(f'output/runs/{ds_id}/*/run.txt.gz'):
    runs[run_name.split('/')[-2]] = pt.io.read_results(run_name)
```

```python
pt.Experiment(
    list(runs.values()),
    topics,
    qrels,
    ["ndcg_cut.10", "ndcg_cut.2", "recip_rank", "recall.1000", "P_10"],
    list(runs.keys())
)
```
